```python
import pandas as pd
import requests
import time
import json
```

```python
df_metadata = pd.read_csv('../../data/metadata-clean.csv')
df_metadata.head()
```

```python
item_ids = df_metadata['item_id'].sort_values().unique()
item_ids
```

```python
url = 'https://steamspy.com/api.php?request=appdetails&appid='
game_tags = {}
last_idx = 28000
```

```python
for i in range(last_idx, len(item_ids)):
    print(f"Processing item {i+1}/{len(item_ids)-last_idx}")
    item_id = item_ids[i]
    # Make the request
    response = requests.get(url + str(item_id))

    # Process the response
    if response.status_code == 200:
        data = response.json()
        game_tags[str(item_id)] = {}
        if isinstance(data['tags'], dict):
            game_tags[str(item_id)]['tags'] = list(data['tags'].keys())
        else:
            game_tags[str(item_id)]['tags'] = data['tags']

        game_tags[str(item_id)]['initialprice'] = data['initialprice']
    else:
        print(f"Error {response.status_code} for {item_id}")
        break

    # Delay for 100ms (0.1 seconds)
    if (i + 1) % 1000 == 0:
        with open(f'../../data/game-tags-{(i+1)/1000}.json', 'w') as f:
            json.dump(game_tags, f)
        game_tags = {}
    elif i == len(item_ids) - 1:
        with open(f'../../data/game-tags-last.json', 'w') as f:
            json.dump(game_tags, f)
        game_tags = {}

    time.sleep(0.1)
```
