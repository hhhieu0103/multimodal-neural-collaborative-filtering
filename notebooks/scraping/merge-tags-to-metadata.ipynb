```python
import pandas as pd
import glob
import json
```

```python
df_metadata = pd.read_csv('../../data/metadata-clean.csv')
df_metadata.head()
```

```python
all_tag_files = glob.glob('../../data/game-tags-*.json')
all_tag_files
```

```python
tags_dict = {}
for tag_file in all_tag_files:
    with open(tag_file, 'r') as f:
        data = json.load(f)
        tags_dict.update(data)

tags_dict
```

```python
ids = tags_dict.keys()
tags = []
prices = []

for id in ids:
    tags.append(tags_dict[id]['tags'])
    prices.append(tags_dict[id]['initialprice'])
```

```python
df_tags = pd.DataFrame({'item_id': ids, 'tags': tags, 'price': prices})
```

```python
df_tags.head()
```

```python
df_metadata['item_id'] = df_metadata['item_id'].astype(str)
df_tags['item_id'] = df_tags['item_id'].astype(str)

df_merged = df_metadata.merge(df_tags, on='item_id', how='left')
df_merged.head()
```

```python
df_merged[['name', 'item_id', 'genres', 'tags', 'price_x', 'price_y']]
```

```python
df_merged['price_y'] = df_merged['price_y'].astype(float)
df_merged['price_y'] = df_merged['price_y'].fillna(0)
df_merged[df_merged['price_x'] != df_merged['price_y']][['name', 'item_id', 'genres', 'tags', 'price_x', 'price_y']]
```

```python
df_merged.drop(columns=['price_x'], inplace=True)
df_merged.rename({'price_y': 'price'}, axis=1, inplace=True)
df_merged.head()
```

```python
df_merged.to_csv('../../data/metadata-clean.csv', index=False)
```
