```python
import pandas as pd
import numpy as np
```

```python
df_interactions = pd.read_csv('data/review-clean.csv')
df_metadata = pd.read_csv('data/metadata-features-extracted.csv')
```

```python
df_interactions.head()
```

```python
df_metadata.head()
```

```python
unique_items_interaction = df_interactions['item_id'].unique()
unique_items_metadata = df_metadata['item_id'].unique()
```

```python
print(f'Number of unique items in interactions: {len(unique_items_interaction)}')
print(f'Number of unique items in metadata: {len(unique_items_metadata)}')
```

```python
interaction_not_metadata = np.setdiff1d(unique_items_interaction, unique_items_metadata)
metadata_not_interaction = np.setdiff1d(unique_items_metadata, unique_items_interaction)
```

```python
print(f'Number of items in interaction but not in metadata: {len(interaction_not_metadata)}')
print(f'Number of items in metadata but not in interactions: {len(metadata_not_interaction)}')
```

```python
all_unique_items = np.union1d(unique_items_interaction, unique_items_metadata)
len(all_unique_items)
```

```python
user_map = {u: i for i, u in enumerate(df_interactions['user_id'].unique())}
item_map = {i: j for j, i in enumerate(all_unique_items)}

df_interactions['user_idx'] = np.array([user_map[u] for u in df_interactions['user_id']])
df_interactions['item_idx'] = np.array([item_map[i] for i in df_interactions['item_id']])
df_metadata['item_idx'] = np.array([item_map[i] for i in df_metadata['item_id']])
```

```python
df_interactions.head()
```

```python
df_metadata.head()
```

```python
df_interactions[['user_id', 'review', 'rating', 'item_id', 'item_idx']].merge(df_metadata[['name', 'item_id', 'item_idx']], on='item_id', how='left')
```

```python
df_interactions.to_csv('data/interaction-synchronized.csv')
df_metadata.to_csv('data/metadata-synchronized.csv')
```
