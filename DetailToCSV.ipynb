```python
import json
import pandas as pd
```

```python
with open('Detail Lists/Detail5545.json', 'r') as file:
    apps = json.load(file)
    
def get_app_data(app):
    id = next(iter(app))
    return app.get(id, {}).get('data', None)

apps = map(get_app_data, apps)
df = pd.DataFrame(apps)

df
```

```python
df.to_csv('abc.csv')
```
