```python
import pandas as pd
```

```python
df_full = pd.read_csv('Data/PricePrediction.csv')
df_full
```

```python
df_features = pd.read_csv('Data/features_extracted.csv')
df_features
```

```python
df_price_missing = pd.read_csv('Data/x_real.csv')
df_price_missing
```

```python
df_price_predicted = pd.read_csv('Data/Missing price predicted.csv')
df_price_predicted
```

```python
df_merged = df_price_missing.drop(columns=['price']).merge(df_price_predicted, on='steam_appid')
df_merged
```

```python
def check_missing_duplicated(df: pd.DataFrame):
    cols_with_nan = df.columns[df.isna().any()].tolist()
    has_dup = df['steam_appid'].duplicated().any()
    print('Missing col:', cols_with_nan)
    if has_dup:
        print('Data Frame has duplicated ID')
```

```python
check_missing_duplicated(df_merged)
```

```python
df_concatenated = pd.concat([df_merged, df_features], ignore_index=True)
df_concatenated
```

```python
check_missing_duplicated(df_concatenated)
```

```python
# Drop encoded columns
df_full.drop(columns=['is_free', 'total_recommendations', 'price', 'supported_languages', 'developers', 'publishers', 'categories', 'genres', 'release_date'], inplace=True)
df_full
```

```python
df_final = df_full.merge(df_concatenated, on='steam_appid')
df_final
```

```python
check_missing_duplicated(df_final)
```

```python
df_final.to_csv('Data/metadata-clean-extracted.csv', index=False)
```
